## Criando nosso banco de dados
- Vamos usar a pasta "bases" com vários arquivos em csv, são nossas tabelas que vamos popular o database

````python
import pandas as pd
import os
````

````python
pasta = os.getcwd() # "getcwd" pega o caminho da pasta atual
pasta
````

````python
arquivos = os.listdir(pasta+"/bases") # "listdir" revela os arquivos da pasta atual, mas se vc der um caminho dentro do método ele fala os arquivos dentro deste caminho que vc especificou
arquivos
````

````python
import sqlite3
con = sqlite3.connect("vendas.db")
cur = con.cursor()
````

### Enviando os arquivos que serão tabelas na pasta "bases" para nosso database "vendas.db"

````python
import pandas as pd

for i in arquivos:
    # Renomear as variaveis, para melhor manunseio
    nome_base = i.replace("olist_", "").replace("_dataset.csv","") # Pegando os arquivos com um nome melhor
    print(nome_base) # Só printando os nomes dos arquivos em pandas caso querermos utilizar, no caso aqui vamos fazer praticamente tudo pelo sql
    
    # Lendo as tabelas com pandas e armazenando em uma variavel, a variavel é usada logo abaixo e convertendo em sql 
    base = pd.read_csv(pasta+"/bases/"+i) 
    
    # Enviando para o database, transformando o arquivo para sql
    base.to_sql(
                nome_base, # nome do arquivo enviado
                con, # conexão atual, fizemos ela lá em cima do código, estamos conectados ao "vendas.db"
                index=False, # tirando a coluna index ao criar as tabelas no db
                if_exists="replace" # tipo de envio, se existir alguma tabela com o mesmo nome vamos dar "replace"
               )
   
````

````python
#IMPORTANTE fazer o commit depois de criar as tabelas para salvar
con.commit()
````

````python
# analisando se deu certo pegando qualquer variavel
# aqui escolhemos a tabela "customers"
cur.execute("SELECT * FROM customers").fetchall() 
````

### Agora que temos todos os arquivos no nosso banco de dados vamos automatizar as consultas criando uma função
- Devolve em arquivo transformado em dataframe
- Renomeia corretamente as colunas
- Mostra o shape
- Mostra os 3 primeiros resultados com head(3)
- Você pode armazenar em uma váriavel para manipular com o pandas pois ela retorna o valor

````python
def execute_query(query):
    # transformando q query no data frame
    resultado = cur.execute(query)
    resultado = pd.DataFrame(resultado)
    
    # Renomeando as colunas
    colunas = [i[0] for i in cur.description]
    if resultado.shape[1] > 0 :
        resultado.columns = colunas
    
    # Printando os resultados, shape, head da coluna para vizualisar e retornando o resultado para ser armazenado em uma variavel quando vc chamar a função 
    print("Shape =" , resultado.shape)
    print("Printando primeiros 3 resultados da busca abaixo")
    display(resultado.head(3))
    return resultado
````

````python
res = execute_query("SELECT * FROM customers")
````

### Temos que tratar a tabela "orders" pois as colunas de envio(as datas) não estão em um formato de data

````python
# Veja que o formato delas estão como object, que é string sendo assim não conseguimos manipular e agrupar por dia anos ou mês, temos que mudar o tipo delas
res_orders.info()
````

### Vamos tratar os dados da base "orders" e por as colunas em de object para data corretamente e depois subir essa base tratada para o sql para substituir a tabela de la pela nossa tabela tratada

````python
# tratando as datas na base "orders"
datas = ["order_purchase_timestamp", "order_approved_at", "order_delivered_carrier_date", "order_delivered_customer_date", "order_estimated_delivery_date"]
for i in datas:
    res_orders[i] = pd.to_datetime(res_orders[i], format="%Y-%m-%d %H:%M:%S")
````

### Trazendo 2 tabelas juntas com o LEFT JOIN
- Vamos juntar a "order_reviews" e a "orders"
- Podemos fazer isso pois elas tem uma coluna em comum 
- Devemos especificar essa coluna na chamada
- A coluna "order_id" existe nas 2 tabelas

### Explicação com Exemplo


### Esquema de Junção
```sql
SELECT coluna1, coluna2, ...
FROM tabela_esquerda
LEFT JOIN tabela_direita ON condição_de_junção;
```
- **Tabela da Esquerda**: `tabela_esquerda` (primeira tabela mencionada no `FROM`)
- **Tabela da Direita**: `tabela_direita` (segunda tabela mencionada no `LEFT JOIN`)

Resumindo:
- A tabela mencionada imediatamente após `FROM` é a tabela da esquerda.
- A tabela mencionada imediatamente após `LEFT JOIN` é a tabela da direita.

Neste caso, o `LEFT JOIN` retornará todos os registros da tabela `customers` (tabela da esquerda) e os registros correspondentes da tabela `orders` (tabela da direita). Se não houver correspondência, os campos da tabela `orders` serão `NULL`.

````python
# Vizualisando antes de juntar
res = execute_query("SELECT * FROM orders")
````

````python
# Juntando elas 
# Perceba que o numero de colunas aumentou e juntou as 2
avaliacoes_e_pedidos = execute_query("SELECT order_reviews.*, orders.* FROM order_reviews \
                                      LEFT JOIN orders \
                                      ON order_reviews.order_id == orders.order_id")
````

### Vamos agora juntar uma terceira tabela 
- Adicionando a "order_items"
- Todas compartilhan o "order_id"

````python
avaliacoes_e_pedidos_e_items = execute_query("""SELECT order_reviews.*, orders.*, order_items.* FROM order_reviews 
                                              LEFT JOIN orders 
                                              ON order_reviews.order_id == orders.order_id 
                                              LEFT JOIN order_items ON 
                                              order_items.order_id == order_reviews.order_id""")
````

````python
# Vendo a média e do preço em relação a avaliação dada usando o pandas
avaliacoes_e_pedidos_e_items.groupby("review_score")["price"].mean()
````
